# Pre processing

```python
df = df.drop("Product", axis = 1)
```

```python
df
```

```python
df = df.join(pd.get_dummies(df.Company))
```

```python
df.drop("Company", axis=1)
```

```python
df = df.join(pd.get_dummies(df.TypeName))
df.drop(columns=['TypeName'], inplace=True)
```

```python
df
```

```python
df.drop(columns=['Company'], axis=1)
```

```python
df['IPS'] = df['ScreenResolution'].str.split(" ").apply(lambda x : 1 if x[0] == 'IPS' else 0)
```

```python
df['ScreenResolution'] = df.ScreenResolution.str.split(" ").apply(lambda x : x[-1])
```

```python
df['Screen_Width'] = df.ScreenResolution.str.split("x").apply(lambda x : x[0])
```

```python
df['Screen_Height'] = df.ScreenResolution.str.split("x").apply(lambda x : x[1])
```

```python
df.drop(columns=['ScreenResolution'], inplace=True)
```

```python
df.head(5)
```

```python
df['CPU_Brand'] = df.Cpu.str.split(" ").apply(lambda x : x[0])
df['CPU_Frequency'] = df.Cpu.str.split(" ").apply(lambda x : x[-1])
```

```python
df.CPU_Brand.value_counts()
```

```python
df.drop(columns=['Cpu'], inplace=True)
```

```python
df['CPU_Frequency'] = df['CPU_Frequency'].str[:-3]
```

```python
df['Ram'] = df['Ram'].str[:-2]
```

```python
df.head()
```

```python
df['Ram'] = df['Ram'].astype('int')
df['CPU_Frequency'] = df['CPU_Frequency'].astype('float')
```

```python
df["Screen_Height"] = df['Screen_Height'].astype('int')
df["Screen_Width"] = df["Screen_Width"].astype('int')
```

```python
df["Memory_Size"] = df["Memory"].str.split(" ").apply(lambda x : x[0])
```

```python
df["Memory_Type"] = df["Memory"].str.split(" ").apply(lambda x : x[1])
```

```python
def mem_to_mb(value):
    if "GB" in value:
        return float(value[:-2])*1000
    if 'TB' in value:
        return float(value[:-2])*1000000
```

```python
df['Memory_Size'] = df['Memory_Size'].apply(mem_to_mb)
```

```python
df.drop("Memory", axis=1)
```

```python
df['Weight'] = df['Weight'].apply(lambda x : x[:-2])
```

```python
df['Weight'] = df['Weight'].astype("float")
```

```python
df['Gpu_Brand'] = df['Gpu'].str.split(" ").apply(lambda x : x[0])
```

```python
df.Gpu_Brand.value_counts()
```

```python
df.drop(columns=['Gpu', 'Memory'], inplace=True)
```

```python
df
```

```python
cpu_categories = pd.get_dummies(df["CPU_Brand"])
cpu_categories.columns = [col + "_Cpu" for col in cpu_categories.columns]
```

```python
os_catagories = pd.get_dummies(df["OpSys"])
```

```python
df = df.join(os_catagories)
```

```python
df.drop(columns=["OpSys"], inplace=True)
```

```python
df
```

```python
df = df.join(cpu_categories)
```

```python
df
```

```python
df.drop(columns=['CPU_Brand'], inplace=True)
```

```python
gpu_catagories = pd.get_dummies(df["Gpu_Brand"])
gpu_catagories.columns = [col + "_GPU" for col in gpu_catagories.columns]
```

```python
df = df.join(gpu_catagories)
```

```python
mem_catagories = pd.get_dummies(df["Memory_Type"])
```

```python
df = df.join(mem_catagories)
```

```python
df.drop(columns=['Gpu_Brand', 'Company', "Memory_Type"], inplace=True)
```

```python
df.columns
```

```python
df = df.replace({True: 1, False: 0})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(18, 15))
sns.heatmap(numerical_df.corr(), annot=True, cmap="YlGnBu")
```

```python
target_correlation = numerical_df.corr()['Price_euros'].apply(abs).sort_values()
```

```python
target_correlation
```

```python
selected_features = target_correlation[-22:].index
```

```python
selected_features = list(selected_features)
```

```python
selected_features 
```

```python
limited_df = df[selected_features]
```

```python
limited_df
```

```python
plt.figure(figsize=(18, 15))
sns.heatmap(limited_df.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x, y = limited_df.drop("Price_euros", axis=1), limited_df['Price_euros']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.12)

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Random Forest Regressor

forest = RandomForestRegressor()

forest.fit(x_train_scaled, y_train)
```

```python
forest.score(x_test_scaled, y_test)
```

```python
y_pred = forest.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(y_pred, y_test)
plt.plot(range(0, 6000), range(0, 6000), c="red")
```
